```python
#Création des dummies variables pour les modalités des variables qualitatives (V160,V161 et V162)

#liste des variables quali
import numpy
lstQuali = [var for var in df.columns[:-1] if df[var].dtype == numpy.object_]
print(lstQuali)
```

```python
dfQualiEncoded = pandas.get_dummies(df[lstQuali],drop_first=True)

print(dfQualiEncoded.info())
```

```python
#liste des variables quantitatives
lstQuanti = [var for var in df.columns[:-1] if df[var].dtype != numpy.object_]
print(lstQuanti)
```

```python
#réunir dummies et quantitatives dans le même data frame
dfNew = pandas.concat([dfQualiEncoded,df[lstQuanti]],axis=1)
print(dfNew.info())
```

```python
#rajouter la variable cible
dfNew['V200'] = df['V200']
print(dfNew.info())
```

```python
############################### Subdivision des données en échantillon test (test) et échantillon d'apprentissage (train)
import sklearn
from sklearn import model_selection
 
train,test = model_selection.train_test_split(dfNew,test_size = round(dfNew.shape[0]*30/100),random_state=1) 
#test_size=30% de 494021 
#random_state fixé à 1 pour pouvoir reproduire l'approche 

print(train.shape,test.shape)
```

```python
############################### Isole les variables prédictives (x) de la variable cible (y) pour chaque échantillon (test et train)

x_train=train.iloc[:,0:dfNew.shape[1]-1]
y_train=train.iloc[:,dfNew.shape[1]-1]
print(y_train.shape, x_train.shape)

x_test=test.iloc[:,0:dfNew.shape[1]-1]
y_test=test.iloc[:,dfNew.shape[1]-1]
print(y_test.shape, x_test.shape)
```

```python
#Première sélection de variables : pour enlever les colonnes de constantes
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold()
selector.fit(x_train)
```

```python
#Liste des variables constantes
print(x_train.columns[selector.variances_ == 0])
```

```python
#Réduction de la base d'apprentissage aux variables sélectionnées
x_train = x_train.iloc[:,selector.get_support()] 
print(x_train.shape) # 270 variables restantes
```

```python
#réduction de la base test aux mêmes variables
x_test = x_test.iloc[:,selector.get_support()] 
print(x_test.shape)
```

```python
###### Regression logistique avec les paramètres initiaux cad solver='lbfgs', et multi_class='multinomial' car la variable cible V200 possède 23 modalités
#Il faut standardiser les données
#importation de l'outil
from sklearn import preprocessing
#instanciation
stds = preprocessing.StandardScaler()
#transformation
z_train = stds.fit_transform(x_train) 
import scipy
print(scipy.stats.describe(z_train,axis=0,ddof=0))
```

```python
#importer la classe LogisticRegression
from sklearn.linear_model import LogisticRegression
#instanciation on garde le solver par défaut et on précise multinomial vu que la variable cible a 23 modalités
lrSkStd = LogisticRegression(solver='lbfgs', multi_class='multinomial', penalty='none',n_jobs=-1)
#lancement des calculs
lrSkStd.fit(z_train,y_train)
```

```python
#affichage des coefficients
print(pandas.DataFrame({"var":x_train.columns,"coef":lrSkStd.coef_[0]}))
```

```python
#correction des coefficients - dé-standardisation
#par les écarts-type utilisés lors de la standardisation des variables 
coefUnstd = lrSkStd.coef_[0] / stds.scale_
#affichage des coefficients corrigés
print(pandas.DataFrame({"var":x_train.columns,"coef":coefUnstd}))
#pour la constante, l'opération est plus complexe
interceptUnStd = lrSkStd.intercept_ + numpy.sum(lrSkStd.coef_[0]*(-stds.mean_/stds.scale_)) 
print(interceptUnStd)
```

```python
#Pour appliquer le modèle sur l’échantillon test, il faut au préalable administrer la même transformation (centrage, réduction) sur ce second dataset en utilisant les paramètres (moyennes, écarts-type) calculés sur l’échantillon d’apprentissage.
#transformation de l'échantillon test
z_test = stds.transform(x_test)
#stat. descriptives
print(scipy.stats.describe(z_test,axis=0,ddof=0))
```

```python
#appliquer la prédiction
predSk = lrSkStd.predict(z_test) 
print(numpy.unique(predSk,return_counts=True))

```

```python
#importation de metrics - utilisé pour les mesures de performances
from sklearn import metrics

#matrice de confusion
cm = metrics.confusion_matrix(y_test,predSk)
print(cm)

```

```python
#taux de succès
acc = metrics.accuracy_score(y_test,predSk) 
print(acc) # 0.9984413586494474

#taux d'erreur
err = 1.0 - acc
print(err)  #0.0015586413505526142
```

```python
#sensibilité (ou rappel)
se = metrics.recall_score(y_test,predSk,average=None) 
print(se) 
```

```python
############################### Determination de meilleurs parametres
#Tester en changeant le solver
#Remarques : 
#multi_class='auto’ selects ‘ovr’ if the data is binary, or if solver=’liblinear’, and otherwise selects ‘multinomial’
#For multiclass problems, only ‘newton-cg’, ‘sag’, ‘saga’ and ‘lbfgs’ handle multinomial loss

#import de la classe
from sklearn.model_selection import GridSearchCV

#instanciation
lr2 = LogisticRegression(n_jobs=-1)


#combinaisons de paramètres à évaluer
#Etant donnée la très grande base de données, l'algorithme de Newton n'est pas l'idéal avec le calcul de la matrice hessienne
#Et liblinear est pour les petites bases de données
#En revanche sag et saga sont plus adaptés pour les grandes bases d'après la documentation
parametres = [{'solver':['sag', 'saga', 'lbfgs'],'multi_class':['ovr','multinomial'],'penalty':['none']}]

grid=GridSearchCV(estimator=lr2,param_grid=parametres,scoring='accuracy',n_jobs=-1)
grille = grid.fit(z_train,y_train)

#résultat pour chaque combinaison
print(pandas.DataFrame.from_dict(grille.cv_results_).loc[:,["params","mean_test_score"]])

#meilleur paramétrage
print(grille.best_params_)  #Le solver lbfgs celui initialement testé mais multi_class='ovr'

#meilleur performance – estimée en interne par validation croisée
print(grille.best_score_) #0.9983806370458191 contre 0.998334 pour celui testé initialement

#prédiction avec le meilleur modèle
y_pred3 = grille.predict(z_test)

#erreur en test
print(1.0 - metrics.accuracy_score(y_test,y_pred3))

```

```python
############################### Regression logistique avec les paramètres optimaux
lr = LogisticRegression(solver='lbfgs', multi_class='ovr', penalty='none',n_jobs=-1)
```

```python
############################### Sélection de variables
from sklearn.feature_selection import SelectFromModel 
selecteur = SelectFromModel(estimator=lr)

#lancer la recherche
sol = selecteur.fit(z_train,y_train)
```

```python
#liste des variables sélectionnées
print(sol.get_support())
```

```python
#réduction de la base d'app. aux var. sélectionnées
#en utilisant le filtre booléen sol.get_support()
z_new_train = z_train[:,sol.get_support()] 
print(z_new_train.shape)
```

```python
#construction du modèle sur les explicatives sélectionnées
modele_selLR = lr.fit(z_new_train,y_train)
```

```python
#réduction de la base test aux mêmes variables
z_new_test = z_test[:,sol.get_support()] 
print(z_new_test.shape)
```

```python
#prédiction du modèle réduit sur l’éch. test
y_pred_selLR = modele_selLR.predict(z_new_test)
print(numpy.unique(y_pred_selLR,return_counts=True))
```

```python
#matrice de confusion
cm = metrics.confusion_matrix(y_test,y_pred_selLR)
print(cm)
```

```python
#évaluation
print(1.0 - metrics.accuracy_score(y_test,y_pred_selLR))
```
